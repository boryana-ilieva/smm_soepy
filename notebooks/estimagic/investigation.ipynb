```python
def process_data(data):
    """Creates additional necessary variables"""
    
    # Determine the education level given years of experience
    data["Educ_Level"] = 0
    data.loc[data["Years_of_Education"] == 11, "Educ_Level"] = 1
    data.loc[data["Years_of_Education"] == 12, "Educ_Level"] = 2

    # Determine the observed wage given period choice
    data["Wage_Observed"] = 0
    data.loc[data["Choice"] == 0, "Wage_Observed"] = data.loc[
        data["Choice"] == 0, "Period_Wage_N"
    ]
    data.loc[data["Choice"] == 1, "Wage_Observed"] = data.loc[
        data["Choice"] == 1, "Period_Wage_P"
    ]
    data.loc[data["Choice"] == 2, "Wage_Observed"] = data.loc[
        data["Choice"] == 2, "Period_Wage_F"
    ]
    
    return data
```

```python
data_sim_base = process_data(data_sim_base)
```

Note:

Illustration why continuation values are always the same
-----------------

Continuation values are the maximum expected value function next period given the current choice. If the only state space components that changes is experience (which is the case), and if and additional year of part-time and of full-time experience results in the same wage next period (which is the case), the continuation value for choosing part-time in a given period is the same as the continuation value of choosing full-time.

```python
log_wage_systematic, non_consumption_utilities = calculate_utility_components(
    model_params, model_spec, states, covariates
)

period = 29
states_period = states[np.where(states[:, 0] == period)]
log_wage_systematic_period = log_wage_systematic[states[:, 0] == period]
non_consumption_utilities_period = non_consumption_utilities[
    states[:, 0] == period
]
```

```python
hours = np.array([0, 18, 38])
draws = draws_emax[29]
num_draws = draws.shape[0]

for state in [idx_pt]:
    
    test= 0.0

    for i in range(num_draws):

        current_max_value_function = -99.0

        for j in range(3):

            wage = np.exp(log_wage_systematic[state] + draws[i][j])
            #print(wage)

            if j == 0:
                consumption_utility = 4 ** model_spec.mu / model_spec.mu
            else:
                consumption_utility = (hours[j] * wage) ** model_spec.mu / model_spec.mu
                
            #print(consumption_utility)

            value_function_choice = (
                consumption_utility * non_consumption_utilities[state][j] + model_spec.delta * emaxs[state][j]
            )
            
            #print(value_function_choice)

            if value_function_choice > current_max_value_function:
                current_max_value_function = value_function_choice
            
            #print(current_max_value_function)

        test += current_max_value_function
        print(test)

    test /= num_draws
```
