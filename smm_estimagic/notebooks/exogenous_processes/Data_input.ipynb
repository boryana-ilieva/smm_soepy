Information on exogenous processes
==============================

The important piece of information regarding the exogenous proceses for the structural estimation are the probabilities with which an exogenous event occurs in a certain period and for an individual with certain characteristics.

We provide a short description of the format in which the informaiton is to be provided. The example files contain randomly generated values. They are to be replaced by values calculated based on our data.

1. Format:
    * .csv file
    * dimensions: number of periods as primary index (1st column) x level of further characteristics as further indices (e.g., second column in exog_partner_info.csv, where education level is the characteristic)
2. Examples
    * Example .csv Files with random values are displayed below and enclosed in the folder

Currently, replacing the randomly generated probabilities in the .csv files is all information we need. As the model grows, we will need to request similar informatio in a comparable format.
